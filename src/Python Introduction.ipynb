```python
# Function Demo

def repeat(s, exclaim):
    """
    This is multiline comment or docstring which describes function
    Return the string 's' repeated 3 times
    If exclaim is True, it adds exclaim marks
    """
    result= s+s+s # result=s*3
    if exclaim:
        result+='!'
    return result

def main():
    # Function call
    print repeat('sumit',True)
    print repeat('- -',False)

if __name__ == '__main__':
    main()

```
